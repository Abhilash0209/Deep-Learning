```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Deep learning/Churn_Modelling.csv")
df.head()
```

```python
x=df.iloc[:,3:-1].values
y=df.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
le2=LabelEncoder()

x[:,1]=le.fit_transform(x[:,1])
x[:,2]=le2.fit_transform(x[:,2])
```

```python
from sklearn.model_selection import train_test_split
xtrain ,xtest ,ytrain ,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
#steps 
#1.intilize the ANN Model
model=Sequential()

#2.pass hiddenlayer
model.add(Dense(units=6,activation="relu"))

#3.Activation Function
model.add(Dense(units=1,activation="sigmoid"))

#4.to establish the connection between layers
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])   

#5.train the model  
model.fit(xtrain,ytrain,batch_size=30,epochs=100)

#6.take prediction
ypred=model.predict(xtest)

#7.set threshhold value
ypred=ypred>0.5

#8.get classification report
print(classification_report(ytest,ypred))
```

```python
#this is  for ml
from sklearn.linear_model import LogisticRegression

logreg=LogisticRegression()

logreg.fit(xtrain,ytrain)

ypred=logreg.predict(xtest)

print(classification_report(ytest,ypred))
```
