```python
def gradient_descent(x,y):
  weight=0.1
  bias=0.1
  learningrate=0.05
  n=len(x)

  epochs=500

  for i in range(epochs):
    #forword propagation

    ypred=x*weight+bias#ypred=x*w+b
    mse=mean_squared_error(y,ypred)

    #backword propagation(step1)

    dw=-(2/n)*sum(x*(y-ypred))
    db=-(2/n)*sum(y-ypred)

    #assign new weight & bias

    weight=weight-learningrate*dw
    bias=bias-learningrate*db

    print(f"iteration:{i},weight:{weight},bias:{bias},MSE:{mse}")
```
